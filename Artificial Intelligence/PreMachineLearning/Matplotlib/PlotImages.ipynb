```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
```

```python
# get image from internet url
url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/0/0b/Cat_poster_1.jpg/1200px-Cat_poster_1.jpg'
img = Image.open(requests.get(url, stream=True).raw)
```

```python
type(img)
```

```python
arr = np.array(img)
```

```python
plt.imshow(arr)
```
