```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100)
```

```python
y = np.sin(x) + 0.2 * x
```

```python
plt.plot(x,y); # semicolon to suppress output
```

```python
plt.plot(x, y, label='sin(x) + 0.2 * x')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('My first plot');
plt.show()
```
