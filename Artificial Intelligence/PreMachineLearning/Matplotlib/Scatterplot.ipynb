```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(100, 2)
plt.scatter(x[:, 0], x[:, 1]) # first column is x, second column is y
plt.show()
```

```python
# Takes the first 50 points and plots them in yellow and move them to the right by 3
x = np.random.randn(200, 2)
x[:50] += 3
y = np.zeros(200)
y[:50] = 1
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show();
```
